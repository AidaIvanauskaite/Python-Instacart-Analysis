### 2. Provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column

```python
hist = instacart['order_hour_of_day'].plot.hist(bins = 24)
```

```python
# The histogram peaks at certain hours, indicating when most orders are placed. Midday or early evening peaks suggest customers shop during lunch or after work. The distribution shape shows if orders are concentrated or spread out. A normal distribution suggests consistent shopping, while a skewed one indicates a preference for specific times like afternoons or evenings.
```
